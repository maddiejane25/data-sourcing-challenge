### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
import os
from datetime import datetime
## Load the NASA_API_KEY from the env file
load_dotenv("example.env")
NASA_API_KEY = os.getenv('NASA_API_KEY')
```

### CME Data

```python
# Set the base URL to NASA's DONKI API:
base_url = "https://api.nasa.gov/DONKI/CME"

# Set the specifier for CMEs:
CME = "CME"

# Search for CMEs published between a begin and end date
startDate = "2013-05-01"
endDate   = "2024-05-01"

# Build URL for CME
CME_url = f"{base_url}?startDate={startDate}&endDate={endDate}&api_key={NASA_API_KEY}"
print(CME_url)
```

```python
# Make a "GET" request for the CME URL and store it in a variable named cme_response
cme_response = requests.get(CME_url)
cme_response
```

```python
# Convert the response variable to json and store it as a variable named cme_json
cme_json = cme_response.json()
```

```python
# Preview the first result in JSON format
first_result = cme_json[0]

# Use json.dumps with argument indent=4 to format data
print(json.dumps(first_result, indent=4))
```

```python
# Convert cme_json to a Pandas DataFrame 
cme_df = pd.json_normalize(cme_json)
cme_df
# Keep only the columns: activityID, startTime, linkedEvents
cme_df_short = cme_df[['activityID', 'startTime', 'linkedEvents']]
cme_df_short
```

```python
# Notice that the linkedEvents column allows us to identify the corresponding GST
# Remove rows with missing 'linkedEvents' since we won't be able to assign these to GSTs
import numpy as np
cme_df_short['linkedEvents'] = cme_df_short['linkedEvents'].replace({None: np.nan})
cme_df_short
cme_df_clean = cme_df_short.dropna(subset=['linkedEvents'])
cme_df_clean
```

```python
# Notice that the linkedEvents sometimes contains multiple events per row
# Write a nested for loop that iterates first over each row in the cme DataFrame (using the index)
# and then iterates over the values in 'linkedEvents' 
# and adds the elements individually to a list of dictionaries where each row is one element 

# Initialize an empty list to store the expanded rows
expanded_rows = []

# Iterate over each index in the DataFrame
for i in cme_df_clean.index:
    activityID = cme_df_clean.loc[i, 'activityID']
    startTime = cme_df_clean.loc[i, 'startTime']
    linkedEvents = cme_df_clean.loc[i, 'linkedEvents']
    # Iterate over each dictionary in the list
    try: 
        for item in linkedEvents:
            x = {'activityID':activityID, 'startTime':startTime, 'linkedEvents':item}
            expanded_rows.append(x)
    except:
        print('missing activity')
        # Append a new dictionary to the expanded_rows list for each dictionary item and corresponding 'activityID' and 'startTime' value
# Create a new DataFrame from the expanded rows
expanded_rows_df = pd.DataFrame(expanded_rows)
expanded_rows_df
```

```python
# Create a function called extract_activityID_from_dict that takes a dict as input such as in linkedEvents
# and verify below that it works as expected using one row from linkedEvents as an example
# Be sure to use a try and except block to handle errors
def extract_activityID_from_dict(input_dict):
    try:
        return input_dict['activityID']
    except KeyError as e:
        print(f"KeyError: {e} - The Key 'activityID' was not found in the dictionary.")
        return None
    except (ValueError, TypeError) as e:
        print(f"Error: {e} - The input was not a valid dictionary.")
        return None
        
```

```python
# Apply this function to each row in the 'linkedEvents' column (you can use apply() and a lambda function)
# and create a new column called 'GST_ActivityID' using loc indexer:
expanded_rows_df.loc[:, 'GST_ActivityID'] = expanded_rows_df['linkedEvents'].apply(lambda x: extract_activityID_from_dict(x))
expanded_rows_df.head()
```

```python
# Remove rows with missing GST_ActivityID, since we can't assign them to GSTs:
clean_cme_df = expanded_rows_df.dropna(subset=['GST_ActivityID'])
clean_cme_df
```

```python
# print out the datatype of each column in this DataFrame:
clean_cme_df.info()
```

```python
# Convert the 'GST_ActivityID' column to string format 
clean_cme_df['GST_ActivityID'] = clean_cme_df['GST_ActivityID'].astype(str)
# Convert startTime to datetime format  
clean_cme_df['startTime'] = pd.to_datetime(clean_cme_df['startTime'])
clean_cme_df.info()
```

```python
# Rename startTime to startTime_CME and activityID to cmeID
clean_cme_df = clean_cme_df.rename(columns={
    'startTime': 'startTime_CME',
    'activityID': 'cmeID'
})
clean_cme_df
# Drop linkedEvents
cme_df_final = clean_cme_df.drop(columns=['linkedEvents'])
cme_df_final
# Verify that all steps were executed correctly
```

```python
# We are only interested in CMEs related to GSTs so keep only rows where the GST_ActivityID column contains 'GST'
# use the method 'contains()' from the str library.  
filtered_cme= cme_df_final[cme_df_final['GST_ActivityID'].str.contains('GST', na=False)]
filtered_cme
```

```python
filtered_cme.info()
```

### GST Data

```python
# Set the base URL to NASA's DONKI API:
base_url = "https://api.nasa.gov/DONKI/GST"

# Set the specifier for Geomagnetic Storms (GST):
GST = "GST"

# Search for GSTs between a begin and end date
startDate = "2013-05-01"
endDate   = "2024-05-01"

# Build URL for GST
GST_url = f"{base_url}?startDate={startDate}&endDate={endDate}&api_key={NASA_API_KEY}"
print(GST_url)
```

```python
# Make a "GET" request for the GST URL and store it in a variable named gst_response
gst_response = requests.get(GST_url)
gst_response
```

```python
# Convert the response variable to json and store it as a variable named gst_json
gst_json = gst_response.json()
# Preview the first result in JSON format
# Use json.dumps with argument indent=4 to format data
print(json.dumps(gst_json[0], indent=4))
```

```python
gst_df = pd.json_normalize(gst_json)
gst_df
```

```python
# Convert gst_json to a Pandas DataFrame  
gst_df = pd.json_normalize(gst_json)

# Keep only the columns: , startTime, linkedEvents
gst_df_short = gst_df[['gstID', 'startTime', 'linkedEvents']]
gst_df_short
```

```python
# Notice that the linkedEvents column allows us to identify the corresponding CME
# Remove rows with missing 'linkedEvents' since we won't be able to assign these to CME
gst_df_short['linkedEvents'] = gst_df_short['linkedEvents'].replace({None: np.nan})
gst_df_short
gst_df_clean = gst_df_short.dropna(subset=['linkedEvents'])
gst_df_clean
```

```python
# Notice that the linkedEvents sometimes contains multiple events per row
# Use the explode method to ensure that each row is one element. Ensure to reset the index and drop missing values.
gst_explode = gst_df_clean.explode('linkedEvents', ignore_index=True).dropna()
gst_explode
```

```python
# Apply the extract_activityID_from_dict function to each row in the 'linkedEvents' column (you can use apply() and a lambda function)
# and create a new column called 'CME_ActivityID' using loc indexer:
gst_explode.loc[:, 'CME_ActivityID'] = gst_explode['linkedEvents'].apply(lambda x: extract_activityID_from_dict(x))
gst_explode
```

```python
# Remove rows with missing CME_ActivityID, since we can't assign them to CMEs:

gst_explode.dropna(subset=['CME_ActivityID'], inplace=True)
gst_explode
```

```python
gst_explode.info()
```

```python
# Convert the 'CME_ActivityID' column to string format 
gst_explode['CME_ActivityID'] = gst_explode['CME_ActivityID'].astype(str)
# Convert the 'gstID' column to string format 
gst_explode['gstID'] = gst_explode['gstID'].astype(str)
# Convert startTime to datetime format  
gst_explode['startTime'] = pd.to_datetime(gst_explode['startTime'])
gst_explode.info()
```

```python
# Rename startTime to startTime_GST 
gst_rename = gst_explode.rename(columns={'startTime':'startTime_GST'})
gst_rename
# Drop linkedEvents
gst_rename.drop('linkedEvents', axis=1, inplace=True)
gst_rename
# Verify that all steps were executed correctly
```

```python
# We are only interested in GSTs related to CMEs so keep only rows where the CME_ActivityID column contains 'CME'
# use the method 'contains()' from the str library.  
filtered_gst= gst_rename[gst_rename['CME_ActivityID'].str.contains('CME', na=False)]
filtered_gst.info()
```

### Merge both datatsets

```python
# Now merge both datasets using 'gstID' and 'CME_ActivityID' for gst and 'GST_ActivityID' and 'cmeID' for cme. Use the 'left_on' and 'right_on' specifiers.
merged_gst_cme = pd.merge(filtered_gst, filtered_cme, left_on=['gstID','CME_ActivityID'], right_on= ['GST_ActivityID','cmeID'], how='outer')
merged_gst_cme
```

```python
# Verify that the new DataFrame has the same number of rows as cme and gst
merged_gst_cme.info()
```

### Computing the time it takes for a CME to cause a GST

```python
# Compute the time diff between startTime_GST and startTime_CME by creating a new column called `timeDiff`.
merged_gst_cme['timeDiff'] = merged_gst_cme['startTime_GST'] - merged_gst_cme['startTime_CME']
merged_gst_cme
```
